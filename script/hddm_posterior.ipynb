```python
import pandas as pd
import numpy as np
import matplotlib as plt
import os
import logging
import arviz as az
```

```python
path_model_infdata = os.path.join('../results/hddm/model_fitted/m_v_t_a.nc')
assert os.path.exists(path_model_infdata), f" nc file not found: {path_model_infdata}"
m_infdata = az.from_netcdf(path_model_infdata)
```

```python
a_condition = ('a(1)','a(2)','a(3)','a(4)')
v_condition = ('v(1)','v(2)','v(3)','v(4)')
t_condition = ('t(1)','t(2)','t(3)','t(4)')
posterior_dict_a = {}
posterior_dict_v = {}
posterior_dict_t = {}
# a
for a in a_condition:
    posterior_dict_a[a] = m_infdata.posterior[a].stack(sample=("chain", "draw")).values

for v in v_condition:
    posterior_dict_v[v] = m_infdata.posterior[v].stack(sample=("chain", "draw")).values

for t in t_condition:
    posterior_dict_t[t] = m_infdata.posterior[t].stack(sample=("chain", "draw")).values

data_posterior_a = pd.DataFrame(posterior_dict_a)
data_posterior_v = pd.DataFrame(posterior_dict_v)
data_posterior_t = pd.DataFrame(posterior_dict_t)

data_posterior_a.to_csv('../results/hddm/data_posterior_a.csv', index=False)
data_posterior_v.to_csv('../results/hddm/data_posterior_v.csv', index=False)
data_posterior_t.to_csv('../results/hddm/data_posterior_t.csv', index=False)
```
